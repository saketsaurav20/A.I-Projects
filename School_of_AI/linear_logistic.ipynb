##Simple LInear regression

##We are starting with 2D linearly separable data . We are using the scikit-learn make_classification method to generate our data and using a helper function to visualize it.

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
def plot_data(X, y, figsize=None):
    if not figsize:
        figsize = (8, 6)
    plt.figure(figsize=figsize)
    plt.plot(X[y==0, 0], X[y==0, 1], 'or', alpha=0.5, label=0)
    plt.plot(X[y==1, 0], X[y==1, 1], 'ob', alpha=0.5, label=1)
    plt.xlim((min(X[:, 0])-0.1, max(X[:, 0])+0.1))
    plt.ylim((min(X[:, 1])-0.1, max(X[:, 1])+0.1))
    plt.legend()
```

```
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
X, y = make_classification(n_samples=1000, n_features=2, n_redundant=0,n_informative=2, random_state=7, n_clusters_per_class=1)
plot_data(X, y)
```

```
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X,y)
print('LR coefficients:', lr.coef_)
print('LR intercept:', lr.intercept_)

plot_data(X, y)
limits = np.array([-2, 2])
boundary = -(lr.coef_[0][0] * limits + lr.intercept_[0]) / lr.coef_[0][1]
plt.plot(limits, boundary, "g-", linewidth=2)
```
